```python
# now for calculating outliers for volume column
# we need to find inter quartile range
iqr = df['Volume'].quantile(0.75) - df['Volume'].quantile(0.25)
iqr
```

```python
#now for outlier
lower_bound = df['Volume'].quantile(0.25) - (1.5*iqr)
upper_bound = df['Volume'].quantile(0.75) + (1.5*iqr)
print(lower_bound)
print(upper_bound)
#
```

```python
# any points outlide lower_bound and upper_bound are considered as outliers
```
