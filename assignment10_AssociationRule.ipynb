##IMP
now here is the main part, this data is single column , all the transactions are saved only at a single column .
So I've converted this data into dataframe df, then created the list of the data for example:(array = [['burgers, meatballs, eggs'], ['pizza, salad']]) Single string list *** . Then I've converted into this format(new_array = [['burgers', 'meatballs', 'eggs'], ['pizza', 'salad']]).Multi string list***. later on every thing is easy

```python

array = df.values.tolist() # creating list
```

```python
new_array = []
for item in array:
    sub_list = []
    for sub_item in item[0].split(','):  # Split on comma only
        sub_list.append(sub_item.strip())  # Strip spaces
    new_array.append(sub_list)

```

Encoding categorical features

```python
encoder = TransactionEncoder()
te_ary = encoder.fit_transform(new_array) # encoding
one_df = pd.DataFrame(te_ary, columns=encoder.columns_) # converting encoded data to dataframe
one_df  # dataframe with encoded information.
```

## Apriori algorithm

It takes encoded transaction data (one_df).
It applies the Apriori algorithm to find all itemsets that appear in at least 5% of the transactions.
It stores these frequent itemsets and their support values in the frequent_itemsets DataFrame.

```python
frequent_itemsets = apriori(one_df, min_support=0.05, use_colnames=True)
frequent_itemsets  # frequent_itemsets contains all the
```

##Association Rules

Interpretation:

This code snippet takes the frequent itemsets identified by the apriori algorithm and generates association rules based on them. It uses the "lift" metric to identify rules where items are purchased together more often than expected by chance. The min_threshold parameter ensures that only rules with a lift value of at least 0.05 are considered.

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.05) # min_treshold is confidence
rules
```

```python
#we have build 6 different rules
```

```python
rules[rules.lift>1.2] # here is how you can apply different features
```

# Interview Questions:

(1) 	What is lift and why is it important in Association rules?

What is Lift?

In association rule mining, lift is a measure of how much more likely two items are to be purchased together than if they were purchased independently. In other words, it quantifies the strength of association between two items.

Why is Lift Important in Association Rules?

1. Identifying Strong Associations
2. Filtering Meaningful Rules
3. Prioritizing Business Decisions

(2)	What is support and Confidence. How do you calculate them?

Support

Support measures the frequency of occurrence of an itemset (a combination of one or more items) in a dataset. It indicates how popular or important an itemset is.

calculation:Support(itemset) = (Number of transactions containing the itemset) / (Total number of transactions)

Confidence

Confidence measures the strength of the association between two itemsets,

Confidence(A -> B) = Support(A and B) / Support(A)

(3) What are some limitations or challenges of Association rules mining?

ans:
Many rules: Too many rules make it hard to find the useful ones.

Rare items: Infrequent items, even if important, might be missed.

Spurious associations: Some rules might appear linked by chance, not true relationships.

Categorical data: Continuous data needs conversion, potentially losing information.

Computational cost: Large datasets or complex rules can make analysis slow
